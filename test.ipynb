```python
import pandas as pd
from pandas.tseries.offsets import MonthEnd
```

```python
industry_ret = pd.read_csv('49_Industry_Portfolios.CSV', skiprows=1, nrows=1160, index_col=0)
industry_ret['caldt'] = pd.to_datetime(industry_ret.index, format='%Y%m', errors='coerce') + MonthEnd(0)
industry_ret = industry_ret.query("caldt >= '1963-07-31'")
industry_ret = industry_ret.reset_index(drop=True)
industry_ret['InNoVaTiOn'] = industry_ret['Softw'] + industry_ret['Telcm'] + industry_ret['ElcEq'] \
                           - industry_ret['Guns'] - industry_ret['Gold'] - industry_ret['Mines']
industry_ret
```

```python
ep_ret = pd.read_csv('Portfolios_Formed_on_E-P.CSV', skiprows=17, nrows=860, index_col=0)
ep_ret['caldt'] = pd.to_datetime(ep_ret.index, format='%Y%m', errors='coerce') + MonthEnd(0)
ep_ret = ep_ret.query("caldt >= '1963-07-31'")
ep_ret = ep_ret.reset_index(drop=True)
ep_ret['EP_HML'] = ep_ret['Hi 10'] - ep_ret['Lo 10']
ep_ret
```

```python
inv_ret = pd.read_csv('Portfolios_Formed_on_INV.CSV', skiprows=17, nrows=716, index_col=0)
inv_ret['caldt'] = pd.to_datetime(inv_ret.index, format='%Y%m', errors='coerce') + MonthEnd(0)
inv_ret = inv_ret.query("caldt >= '1963-07-31'")
inv_ret = inv_ret.reset_index(drop=True)
inv_ret['INV_HML'] = inv_ret['Hi 10'] - inv_ret['Lo 10']
inv_ret
```

```python
op_ret = pd.read_csv('Portfolios_Formed_on_OP.CSV', skiprows=24, nrows=726, index_col=0)
op_ret['caldt'] = pd.to_datetime(op_ret.index, format='%Y%m', errors='coerce') + MonthEnd(0)
op_ret = op_ret.query("caldt >= '1963-07-31'")
op_ret = op_ret.reset_index(drop=True)
op_ret['OP_HML'] = pd.to_numeric(op_ret['Hi 10']) - pd.to_numeric(op_ret['Lo 10'])
op_ret
```

```python
url = 'http://diether.org/prephd/10-port_mom_ew.csv'
port = pd.read_csv(url,parse_dates=['caldt'])
port
```

```python
df = pd.merge(pd.merge(pd.merge(pd.merge(port, industry_ret, on='caldt'), inv_ret[['caldt', 'INV_HML']], on = 'caldt'),ep_ret[['caldt', 'EP_HML']], on = 'caldt'), op_ret[['caldt', 'OP_HML']], on = 'caldt')
df
```

```python
from statsmodels.formula.api import ols
ols('p4 ~ 1 + INV_HML + EP_HML + OP_HML + InNoVaTiOn', data = df).fit().summary()
```
