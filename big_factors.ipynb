```python
from class.FamaData import FamaData
import pandas as pd
import numpy as np
```

```python
fd = FamaData('data/', end_date = '2022-09-30')
factor_port = fd.get_49_industry_data()
cols = factor_port.columns[:-2] # Ignore caldt and other
print(cols)
big_factors = pd.DataFrame(columns = ['caldt'])
big_factors['caldt'] = factor_port['caldt']
# Avoid double-counting factors
for i, col_i in enumerate(cols):
    for j, col_j in enumerate(cols):
        if i < j:
            print(cols[i], cols[j])
            big_factors[cols[i].strip() + 'Minus' + cols[j].strip()] = factor_port[cols[i]] - factor_port[cols[j]]
big_factors.head(5)
big_factors.to_csv('exported/big_factors.csv', index = False)
```

```python
corr_matrix = big_factors.corr()
corr_matrix = corr_matrix.where(np.triu(np.ones(corr_matrix.shape)).astype(bool))
```
