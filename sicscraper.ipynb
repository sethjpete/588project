```python
import io
mapper = {}
with open ('Siccodes49.txt', 'r') as f:
    last_section = ""
    for line in f:
        if line.startswith('          '):
            x = int(line.split('          ')[1].split(' ')[0].split('-')[0])
            y = int(line.split('          ')[1].split(' ')[0].split('-')[1])
            mapper[x] = last_section
            mapper[y] = last_section
        else:
            try:
                last_section = line.split("  ")[1].strip()
                print(last_section)
            except:
                ;
mapper
```

```python
import json

with open('sicmap.json', 'w') as fp:
    json.dump(mapper, fp)
```
